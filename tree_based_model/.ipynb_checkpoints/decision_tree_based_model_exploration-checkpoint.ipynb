# 5. å†³ç­–æ ‘çš„å®ç°
sklearné‡‡ç”¨CARTç®—æ³•æ¥è®­ç»ƒå†³ç­–æ ‘ï¼ˆæœ‰æ—¶ä¹Ÿå«â€œç”Ÿé•¿æ ‘â€ï¼‰ã€‚CARTç®—æ³•çš„æ€æƒ³å¾ˆç›´è§‚ï¼šç”¨ç‰¹å¾$k$å’Œä¸€ä¸ªé˜ˆå€¼$t_k$æ¥å°†è®­ç»ƒæ ·æœ¬åˆ†æˆä¸¤ä¸ªå­é›†ï¼Œé€‰æ‹©$(k,t_k)$çš„æ–¹æ³•æ˜¯ï¼Œå°±æ˜¯è¦ä½¿å¾—å­é›†çš„çº¯å‡€åº¦æœ€é«˜ï¼Œçº¯å‡€åº¦å¯é€šè¿‡Giniç³»æ•°å’Œä¿¡æ¯ç†µæ¥è¡¡é‡ã€‚

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

tree_clf = DecisionTreeClassifier(criterion='gini', 
                                  max_depth=2, 
                                  min_samples_leaf=2,
                                  min_samples_split=2,
                                  max_features=2)
tree_clf.fit(X, y)

y_pred = tree_clf.predict(X)

tree_clf.score(X,y)
```

# 6. Bagging ä¸éšæœºæ£®æ—

éšæœºæ£®æ—å±äºé›†æˆå­¦ä¹ æ–¹æ³•ä¸­çš„ä¸€ç§ï¼Œé›†æˆå­¦ä¹ ä¸»è¦æ€æƒ³æ˜¯ï¼Œé‡‡ç”¨å¤šç§ç®—æ³•è®­ç»ƒå¾—åˆ°ä¸°å¯Œçš„åˆ†ç±»å™¨ï¼Œç„¶åå°†å®ƒä»¬çš„é¢„æµ‹ç»“æœæ•´åˆåœ¨ä¸€èµ·ï¼Œå¾—åˆ°ä¸€ä¸ªæ›´å¥½çš„é¢„æµ‹ç»“æœã€‚æ¢å¥è¯è¯´å°±æ˜¯ï¼Œâ€œä¸‰ä¸ªè‡­çš®åŒ ï¼Œé¡¶ä¸ªè¯¸è‘›äº®â€ã€‚å°†å¤šä¸ªï¼ˆå¼±ï¼‰åˆ†ç±»å™¨çš„é¢„æµ‹ç»“æœæ•´åˆçš„æ–¹æ³•ï¼Œæœ€ç®€å•çš„å°±æ˜¯é€šè¿‡å¤šæ•°æŠ•ç¥¨å’Œå–å¹³å‡ã€‚ä¸‹å›¾æ˜¯é‡‡ç”¨å¤šæ•°æŠ•ç¥¨çš„æ–¹æ³•å¾—åˆ°æœ€ç»ˆé¢„æµ‹ç»“æœçš„ç¤ºæ„å›¾

## Voting Classifier
![image.png](./images/voting_classifier.png)

æœ‰ç‚¹è®©äººåƒæƒŠçš„æ˜¯ï¼Œé‡‡ç”¨ä¸Šè¿°æ–¹æ³•å¾—åˆ°çš„æœ€ç»ˆåˆ†ç±»å™¨ï¼Œå‡†ç¡®åº¦ä¸€èˆ¬è¦æ¯”é‡‡ç”¨å•ä¸ªåˆ†ç±»å™¨å‡†ç¡®ç‡è¦é«˜ã€‚åœ¨scikit-learnä¸­ä¹Ÿæœ‰å¯¹è¿™ç§å¤šæ•°æŠ•ç¥¨æ³•å®ç°é›†æˆå­¦ä¹ çš„å®ç°ï¼Œå³VotingClassifier()ã€‚ä¸‹é¢æ˜¯å°†Decision Treeï¼ŒLogistics regressionä»¥åŠSVMé‡‡ç”¨å¤šæ•°æŠ•ç¥¨çš„æ–¹æ³•ç»„æˆçš„åˆ†ç±»å™¨ï¼Œä»ç»“æœå¯çœ‹åˆ°ï¼Œå¯¹è¯¥æ•°æ®é›†ï¼Œé›†æˆå­¦ä¹ çš„åˆ†ç±»å‡†ç¡®åº¦æ›´é«˜ã€‚

```python
# é‡‡ç”¨å¤šä¸ªæ¥è‡ªä¸åŒç®—æ³•çš„åˆ†ç±»æ¨¡å‹ä»¥baggingçš„æ–¹å¼åšensemble learningï¼š
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples=3000, noise=0.9, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
dt_clf = DecisionTreeClassifier(random_state=42)
lr_clf = LogisticRegression(random_state=42)
svm_clf = SVC(random_state=42)

voting_clf = VotingClassifier(estimators=[('lr', lr_clf),
                                          ('dt', dt_clf),
                                          ('svm', svm_clf)],
                             voting='hard')
voting_clf.fit(X_train, y_train)

for clf in (dt_clf, lr_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
    
```

## Bagging
Bagging (short for bootstrap aggregating) ä¹Ÿæ˜¯ä¸€ç§é›†æˆå­¦ä¹ ï¼ˆensemble learningï¼‰çš„æŠ€æœ¯ã€‚ä¸Šé¢è®²çš„VotingClassifieræ˜¯ç”¨ä¸åŒçš„ç®—æ³•å¾—åˆ°çš„å‡ ä¸ªæ¨¡å‹è¿›è¡Œæ•´åˆï¼Œäº‹å®ä¸Šï¼Œæˆ‘ä»¬å¯ä»¥ç”¨åŒä¸€ç§ç®—æ³•åœ¨ä¸åŒçš„è®­ç»ƒé›†å­é›†ä¸Šè®­ç»ƒå¾—åˆ°ä¸åŒçš„å¼±åˆ†ç±»å™¨ï¼Œç„¶åå†å°†å¤šä¸ªå¼±å­¦ä¹ å™¨çš„é¢„æµ‹ç»“æœç»“åˆèµ·æ¥ï¼Œä»è€Œé™ä½é¢„æµ‹è¯¯å·®ã€‚æ¯ä¸ªå¼±å­¦ä¹ å™¨æ˜¯åœ¨è®­ç»ƒé›†çš„å­é›†ä¸Šå»ºç«‹äº†ã€‚ä¸‹å›¾å¯ä»¥å¾ˆå½¢è±¡çš„è¯´æ˜ï¼š

![](./images/bagging.png)

baggingé€šè¿‡å¯¹è®­ç»ƒé›†è¿›è¡Œæœ‰æ”¾å›çš„é‡‡æ ·æ¥ç”Ÿæˆä¸åŒçš„å­é›†ï¼Œä»è€Œåœ¨å„å­é›†ä¸Šè®­ç»ƒä¸åŒçš„å¼±å­¦ä¹ å™¨ï¼Œä¹Ÿå°±æ˜¯è¯´ï¼Œå¯¹æ¯ä¸ªå¼±å­¦ä¹ å™¨æ¥è¯´ï¼Œbaggingå…è®¸å¯¹ç›¸åŒçš„æ ·æœ¬é‡‡æ ·å¤šæ¬¡ã€‚è‹¥æ¯ä¸ªå¼±å­¦ä¹ å™¨éƒ½å¯ä»¥å¯¹åˆ†ç±»é—®é¢˜äº§ç”Ÿé¢„æµ‹æ¦‚ç‡ï¼Œé‚£ä¹ˆbaggingçš„æœ€ç»ˆé¢„æµ‹æ¦‚ç‡ä¸ºå®ƒä»¬çš„åŠ æƒå¹³å‡ï¼Œè¿™å«â€œ**soft voting**ï¼›è‹¥ä¸èƒ½é¢„æµ‹æ¦‚ç‡ï¼ˆå¦‚SVM åˆ†ç±»å™¨ï¼‰ï¼Œé‚£ä¹ˆé‡‡ç”¨å¤šæ•°æŠ•ç¥¨çš„æ–¹å¼æ¥ç”Ÿæˆæœ€ç»ˆçš„é¢„æµ‹ç»“æœï¼Œè¿™å«**â€hard votingâ€œ**

Scikit-learné›†æˆäº†é‡‡ç”¨bootstré‡‡æ ·çš„baggingåˆ†ç±»å™¨ï¼Œå³BaggingClassifier()ï¼Œä»¥ä¸‹æ˜¯é‡‡ç”¨å¤šä¸ªå¼±åˆ†ç±»å™¨ï¼ˆå†³ç­–æ ‘ï¼‰å®ç°baggingæ–¹æ³•çš„ä»£ç ï¼š

```python
from sklearn.ensemble import BaggingClassifier 
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(DecisionTreeClassifier(), 
                            n_estimators=500, 
                            max_samples=100, 
                            bootstrap=True, 
                            n_jobs=-1)

bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```
